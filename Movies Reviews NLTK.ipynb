2\Bag of Words (BOW): is a method to extract features from text documents. These features can be used for training machine learning algorithms. It creates a vocabulary of all the unique words occurring in all the documents in the training set.

```python
useless_words = nltk.corpus.stopwords.words('english') + list (string.punctuation)
filtered_words = [word for word in movie_reviews.words() if not word in useless_words]
```

3\ Counter frequencies : Counting the item frequencies of a list returns a dictionary-like object which displays the number of times each element appears in the list.

```python
from collections import Counter
word_counter = Counter(filtered_words)
word_counter
```

```python
most_common_words = word_counter.most_common()[:10]
most_common_words
```

# SENTIMENT ANALYSIS MODEL

```python
#build features
def build_bag_of_words(words):
    return  {
        word:True for word in words \
        if not word in useless_words}
```

```python
neg_reviews = []
for fileid in movie_reviews.fileids('neg'):
    # We get all the words in that file.
    words = movie_reviews.words(fileid)
    # Then we use the function we wrote earlier to create word features in the format nltk expects.
    neg_reviews.append((build_bag_of_words(words), "negative"))
```

```python
neg_reviews[0]
```

```python
pos_reviews = []
for fileid in movie_reviews.fileids('pos'):
    words = movie_reviews.words(fileid)
    pos_reviews.append((build_bag_of_words(words), "positive"))
```

```python
pos_reviews[0]
```

# TRAIN THE MODEL

```python
from nltk.classify import NaiveBayesClassifier
```

```python
len(pos_reviews)
```

```python
train_set = neg_reviews[:750] + pos_reviews[:750]
test_set =  neg_reviews[750:] + pos_reviews[750:]
```

```python
classifier = NaiveBayesClassifier.train(train_set)
```

```python
accuracy = nltk.classify.util.accuracy(classifier, test_set)
print(accuracy * 100)
```
